# Data Analysis

```python
import pandas as pd
bd = pd.read_csv('cars.csv', encoding= 'ISO-8859-1')
```

```python
# Dataframe summary: only numeric columns
bd.describe(include ="all")
```

```python
summary = bd.describe()
```

```python
percentiles = summary.loc[['25%', '75%']]
```

```python
print("25th and 75th Percentiles:")
print(percentiles)
```

```python
print("\nInterpretation:")
for column in percentiles.columns:
    print(f"- For {column}:")
    print(f"  25% of the data lies below {percentiles.loc['25%', column]}")
    print(f"  75% of the data lies below {percentiles.loc['75%', column]}, "
          f"meaning the top 25% are above this value.\n")
```
